```python
import pandas as pd
import numpy as np
import matplotlib as mplt
```

```python
df = pd.read_csv('analysis/p2p/LoanStats3a.csv', parse_dates=True, infer_datetime_format=True, header=1)
```

```python
df
```

```python
import re
```

```python
begin_regex = re.compile(r'^')
df = df.assign(issue_d = lambda x : pd.to_datetime(x['issue_d'].str.replace(begin_regex, '01-'), format="%d-%b-%Y"))
```

```python
df['issue_d']
```

```python
df.plot(x = 'issue_d', y = 'funded_amnt')
```
